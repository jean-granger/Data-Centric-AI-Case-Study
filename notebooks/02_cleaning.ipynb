```python
import os, sys
sys.path.append(os.path.abspath(".."))  

import pandas as pd
from src.preprocessing import basic_impute, summarize_missing

# Load the raw dataset
df = pd.read_csv("../data/raw/heart.csv")
df.head()
```

```python
df_imputed = basic_impute(df.copy(), numeric_strategy='median')
display(summarize_missing(df_imputed))
```

```python
cat_cols = df_imputed.select_dtypes(include=['object','category']).columns.tolist()
df_encoded = pd.get_dummies(df_imputed, columns=cat_cols, drop_first=True)

df_encoded.to_csv("../data/processed/heart_processed.csv", index=False)
print("Saved processed dataset to data/processed/heart_processed.csv")
df_encoded.shape
```

#Cleaning Results
- Applied median imputation (no missing values to fix in this dataset)
- Checked for categorical variables (none found here) 
- Saved cleaned dataset as `data/processed/heart_processed.csv`.
